### Q 8. Convert the column "Reviews" to numeric data type and check the presence of outliers in the column and handle the outliers using a transformation approach.(Hint: Use log transformation)

```python
# we change the datatype of the reviews column to int by using .astype(int) function.
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
# .boxplot() is used to check the outliers present in the reviews data.
sns.boxplot(y = df_copy['Reviews'],data = df_copy)
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer

# using log transformation we convert the data range to handle the outliers.
log_transformer = FunctionTransformer(np.log1p)

log_transformer
```

```python
# .fit_transform is used to fit the data inside the boxplot without outseeding into outliers for the reviews column
df_copy['Reviews_log'] = log_transformer.fit_transform(df_copy[['Reviews']])
```

```python
df_copy['Reviews_log'].head()
```

```python
df_copy['Reviews_log'].describe()
```

```python
plt.boxplot(df_copy['Reviews_log'])

# mentioning the min, max, median for the quantile (q1, q3) were to visually represent the data clearly to get details.
plt.text(x = 1.1, y = df_copy['Reviews_log'].min(), s ='min')
plt.text(x = 1.1, y = df_copy.Reviews_log.quantile(0.25), s ='Q1')
plt.text(x = 1.1, y = df_copy['Reviews_log'].median(), s ='median (Q2)')
plt.text(x = 1.1, y = df_copy.Reviews_log.quantile(0.75), s ='Q3')
plt.text(x = 1.1, y = df_copy['Reviews_log'].max(), s ='max')

plt.title('Boxplot of Reviews_log')
plt.ylabel('Reviews')

plt.show()
```
