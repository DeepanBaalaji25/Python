### Q3. Display the percentage distribution of the data in each year using the pie chart.

```python
# values_counts() function is used to find the total values in each year
df['year'].value_counts()
```

```python
# assigning x as variable and implying value_counts() of year into it.
x = df['year'].value_counts()

# assigning y as variable and adding years into it.
y = ['2001','2002','2003','1998','1999','2000','2004','2005','2006','2007','2008','2009','2010','2011','2012']

# using .pie() to create a peiplot to visually represent the data of year clearly, for a better view with percentage.
plt.pie(x, labels = y , autopct = '%.1f%%',pctdistance = 0.75,shadow = True,radius = 1.5)
plt.show()
```

### Q4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.
