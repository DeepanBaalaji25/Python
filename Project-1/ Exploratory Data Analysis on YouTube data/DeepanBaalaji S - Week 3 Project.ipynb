Q6. Create a new column as 'published_month' using the column published_at (display the months only)

```python
# By using published_at column we can easily create a new column named published_month with only month by using .dt.month() function
df_copy['published_month'] = df_copy['published_at'].dt.month
```

Q7. Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb'
and so on.....

```python
#We can create a dictionary to convert months as name and store it in a variable and apply it to the dataframe.
months_name = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}

# The converted one is replaced to the published_month by using .replace() function
df_copy['published_month'] = df_copy['published_month'].replace(months_name)
```

```python
# Now the changed value will be displayed in the data by calling it
df_copy
```

Q8. Find the number of videos published each month and arrange the months in a decreasing order based

Q10. Find the top10 channel names having the highest number of videos in the dataset and the bottom10
having lowest number of videos.


```python
# Using groupby function for channel_title & video_id and use count() function to fnd the total count of the videos of each channel.
# Now sort them using .sort_values and give ascending=false, so that we get all highest numebr of videos in the channels
# to get top 10 we use .head(10) for the first top 10 channels of the dataset.

top_10 = df_copy.groupby('channel_title')['video_id'].count().sort_values(ascending=False).head(10)
print('The top 10 channels having highest number of videos are:','\n',top_10)
```

```python
# Using groupby function for channel_title & video_id and use count() function to fnd the total count of the videos of each channel.
# Now sort them using .sort_values and give ascending=False, so that we get all Lowest number of videos in the channels
# to get bottom 10 we use .tail(10) for the bottom 10 channels of the dataset.

bottom_10 = df_copy.groupby('channel_title')['video_id'].count().sort_values(ascending=False).tail(10)
print("The bootom 10 channels having lowest number of videos are:",'\n',bottom_10)
```

```python
# Use .corr() function to find the correlation between the coulmns and put a graph using scatter plot
df_copy.corr(numeric_only=True)
```

```python
# To get the comparision of relationship between 2 numerical values of (view_count & dislikes) we use scatter plot. 
sns.scatterplot(data=df_copy,x='view_count',y='dislikes')
plt.show()
```

By observing the plot we can stat that when the view_count of the video increases the rate of dislike is randomized which we can say for a lesser view_count the dislike count is high for few videos and for a higher view_count the dislike is approximately to its viewers rate of the viewers count

```python
# by using seaborns and heatmap() function we can create it with the numerical columns
sns.heatmap(df_copy.corr(numeric_only=True),annot=True,fmt='.2f')
plt.show()
```
